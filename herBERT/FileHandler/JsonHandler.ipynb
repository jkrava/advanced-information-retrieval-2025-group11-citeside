```python
from pathlib import Path
import webbrowser
import html


def open_text_in_browser(text, filename="paper.html"):
    safe_text = html.escape(text)
    html_content = f"""
    <!DOCTYPE html>
    <html>
    <head>
        <meta charset="utf-8">
        <title>Paper Full Text</title>
        <style>
            body {{
                margin: 40px auto;
                max-width: 900px;
                font-family: Georgia, serif;
                line-height: 1.6;
                font-size: 16px;
                background: #ffffff;
                color: #111;
            }}
            pre {{
                white-space: pre-wrap;
            }}
        </style>
    </head>
    <body>
        <pre>{safe_text}</pre>
    </body>
    </html>
    """

    path = Path(filename).resolve()
    path.write_text(html_content, encoding="utf-8")

    webbrowser.open(path.as_uri())
```
