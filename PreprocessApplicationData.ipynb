````python
from grobid_client.grobid_client import GrobidClient

client = GrobidClient()

#pdfs_path = "Datasets/Application Data/Pompei"
pdfs_path = "Datasets/Application Data/Thermopylen"

#output_path = "Datasets/Application Data/Pompei-Grobid-Output"
output_path = "Datasets/Application Data/Thermopylen-Grobid-Output"


# client.process(
#     service: str - the service to call (e.g., "processFulltextDocument")
#     input_path: str - path to input PDFs
#     output: str - path to output directory
#     n: int - number of parallel processes (default: 10)
#     generateIDs: bool - generate document IDs (default: False)
#     consolidate_header: bool - consolidate header (default: True)
#     consolidate_citations: bool - consolidate citations (default: False)
#     include_raw_citations: bool - include raw citations (default: False)
#     include_raw_affiliations: bool - include raw affiliations (default: False)
#     tei_coordinates: bool - include TEI coordinates (default: False)
#     segment_sentences: bool - segment sentences (default: False)
#     force: bool - force processing (default: True)
#     verbose: bool - verbose output (default: False)
#     flavor: str - processing flavor (default: None)
#     json_output: bool - output as JSON (default: False)
#     markdown_output: bool - output as Markdown (default: False))
##

client.process("processFulltextDocument",
                input_path=pdfs_path,
                output=output_path,
                n = 2,
                json_output=True)

         
````
